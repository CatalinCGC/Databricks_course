```python
# Read date as input and get the day of the week (Sun, Mon, Tue)
# Create input date widget

dbutils.widgets.text("input_date", "")
```

```python
# Read value from widget
_input_date = dbutils.widgets.get("input_date")
print(_input_date)
```

```python
# FOrmat the input date to extract the day of the week
spark.conf.set("spark.sql.legacy.timeParserPolicy","LEGACY")

_input_day = spark.sql(f"""
                       select date_format(to_timestamp('{_input_date}',"yyyy-MM-dd"), 'E')
                       """).collect()

display(_input_day)

```

```python
# Export the day of the week using taskValues in next task

dbutils.jobs.taskValues.set("input_day", _input_day[0][0])
```
