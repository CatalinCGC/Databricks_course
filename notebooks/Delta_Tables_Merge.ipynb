```sql
-- Merge data to update existing records, insert new records and delete missing records

MERGE INTO dev.bronze.emp AS target
USING dev.bronze.emp_updates AS source
ON target.emp_id = source.emp_id
WHEN MATCHED THEN
UPDATE SET *
WHEN NOT MATCHED THEN
INSERT *
WHEN NOT MATCHED BY SOURCE THEN
DELETE
;
```

```sql
-- Soft delete (IS_ACTIVE = N)
MERGE INTO dev.bronze.emp AS target
USING dev.bronze.emp_updates AS source
ON target.emp_id = source.emp_id
WHEN MATCHED THEN
UPDATE SET 
  target.salary = source.salary,
  target.is_active = 'Y'
WHEN NOT MATCHED THEN
INSERT (
    emp_id,
    emp_name,
    dept_code,
    salary,
    is_active
)
VALUES (
    source.emp_id,
    source.emp_name,
    source.dept_code,
    source.salary,
    'Y'
)
WHEN NOT MATCHED BY SOURCE  AND target.is_active = 'Y' THEN
UPDATE SET is_active = 'N'
;
```
