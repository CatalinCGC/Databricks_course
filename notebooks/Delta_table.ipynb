# Temporary and permanent views

```sql
CREATE TEMPORARY VIEW emp_tmp_vw AS SELECT * FROM dev.bronze.emp where dept_code = 'D101'
```

```sql
select * from emp_tmp_vw
```

```sql
CREATE VIEW dev.bronze.emp_vw AS SELECT * FROM dev.bronze.emp where dept_code = 'D102'
```

```sql
select * from dev.bronze.emp_vw
```

```sql
CREATE OR REPLACE VIEW dev.bronze.emp_vw AS SELECT * FROM dev.bronze.emp where dept_code = 'D101'
```

```sql
select * from dev.bronze.emp_vw
```

```sql
CREATE TABLE dev.bronze.emp_ctas AS SELECT * FROM dev.bronze.emp;
select * from dev.bronze.emp_ctas;
```

```sql
describe extended dev.bronze.emp_ctas;
```

```sql
describe history dev.bronze.emp_ctas;
```

# Deep clone (includes metadata & data)

```sql
CREATE TABLE dev.bronze.emp_dc DEEP CLONE dev.bronze.emp;
select * from dev.bronze.emp_dc;
```

```sql
describe history dev.bronze.emp_dc;
```

# Shallow clone - metadata only

```sql
CREATE TABLE dev.bronze.emp_sc SHALLOW CLONE dev.bronze.emp;
select * from dev.bronze.emp_sc;
```

```sql
describe extended dev.bronze.emp_sc
```

```sql
describe history dev.bronze.emp_sc
```
